# Simple Linear Regression On Salary Dataset

```python
# importing libraries and loading dataset 
import pandas as pd
import numpy as np
df=pd.read_csv('C:/Users/SONY/Documents/CSV_Files/salary.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
x=df[['YearsExperience']].values
y=df[['Salary']].values
```

```python
from sklearn import preprocessing
x=preprocessing.StandardScaler().fit(x).transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train,color='blue')
plt.plot(x_train,reg.predict(x_train),color='black')
plt.title('Training data plot')
plt.xlabel('Year of Experince')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,reg.predict(x_train),color='black')
plt.title('Training data plot')
plt.xlabel('Year of Experince')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn import metrics
MAE=metrics.mean_absolute_error(y_test,y_pred)
MSE=metrics.mean_squared_error(y_test,y_pred)
RMSE=np.sqrt(metrics.mean_absolute_error(y_test,y_pred))
r2_score=metrics.r2_score(y_test,y_pred)
print('mean_absolute_error: ',MAE)
print('mean_squarred_error: ',MSE)
print('root_mean_absolute_error: ',RMSE)
print('r2_score: ',r2_score)
```

Result- By applying Simple Linear Regression on Salary dataset, we predict that salary is directly proportional to Years of experince of that person with 96% accuracy.
